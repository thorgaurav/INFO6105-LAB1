#### Create 3 col. for numbers for 1%, 5% and 10% missing data - Column selected - ShapeFactor1


```python
df_drybean_ShapeFactor1 = df_drybean[['ShapeFactor1']]
df_drybean_ShapeFactor1['ShapeFactor1_copy_1_percent'] = df_drybean[['ShapeFactor1']]
df_drybean_ShapeFactor1['ShapeFactor1_copy_5_percent'] = df_drybean[['ShapeFactor1']]
df_drybean_ShapeFactor1['ShapeFactor1_copy_10_percent'] = df_drybean[['ShapeFactor1']]
df_drybean_ShapeFactor1['Index'] = range(1, len(df_drybean_ShapeFactor1) + 1)
# df_drybean_ShapeFactor1['sample_num'] = df_drybean['sample_num']/1000
df_drybean_ShapeFactor1
```

```python
def get_percent_missing(dataframe):
    
    percent_missing = dataframe.isnull().sum() * 100 / len(dataframe)
    missing_value_df = pd.DataFrame({'column_name': dataframe.columns,
                                     'percent_missing': percent_missing})
    return missing_value_df
```

```python
print(get_percent_missing(df_drybean_ShapeFactor1))
```

```python
def create_missing(dataframe, percent, col):
    dataframe.loc[dataframe.sample(frac = percent).index, col] = np.nan
```

```python
create_missing(df_drybean_ShapeFactor1, 0.01, 'ShapeFactor1_copy_1_percent')
create_missing(df_drybean_ShapeFactor1, 0.05, 'ShapeFactor1_copy_5_percent')
create_missing(df_drybean_ShapeFactor1, 0.1, 'ShapeFactor1_copy_10_percent')
```

```python
from math import sqrt
variance_1_percent = (imputed_number_df['diff_1_percent'] ** 2)/(len(imputed_number_df)-1)
variance_5_percent = (imputed_number_df['diff_5_percent'] ** 2)/(len(imputed_number_df)-1)
variance_10_percent = (imputed_number_df['diff_10_percent'] ** 2)/(len(imputed_number_df)-1)
print("Variance for 1 percent missing data is  ",variance_1_percent)
print("Variance for 5 percent missing data is  ",variance_5_percent)
print("Variance for 10 percent missing data is  ",variance_10_percent)

```

```python
bias_1_percent = (acutual - predicted_1_percent) / N
bias_5_percent = (acutual - predicted_5_percent) / N
bias_10_percent = (acutual - predicted_10_percent) / N
print("Bias for 1 percent missing data is = ",bias_1_percent)
print("Bias for 5 percent missing data is = ",bias_5_percent)
print("Bias for 10 percent missing data is = ",bias_10_percent)
```

```python
#RSME Calculation
ERR_1_percent = imputed_number_df['diff_1_percent'].sum(axis=0) 
ERR_5_percent = imputed_number_df['diff_5_percent'].sum(axis=0) 
ERR_10_percent = imputed_number_df['diff_10_percent'].sum(axis=0) 

```

```python
print("RMSE for 1 percent missing data is", sqrt(ERR_1_percent))
print("RMSE for 5 percent missing data  is", sqrt(ERR_5_percent))
print("RMSE for 10 percent missing data  is", sqrt(ERR_10_percent))
```
